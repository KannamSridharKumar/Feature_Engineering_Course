```python
import jenkspy
import random
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (12, 4)
# plt.rcParams["axes.grid"] = True
plt.rcParams.update({"font.size": 14})
```

```python
feat = np.concatenate(
    [
        np.random.randint(20000, 80000, 600),
        np.random.randint(80000, 120000, 200),
        np.random.randint(120000, 180000, 150),
        np.random.randint(180000, 250000, 50),
    ]
)

feat = np.concatenate(
    [
        np.random.normal(50, 10, 1000),
        np.random.uniform(90, 110, 1000),
        np.random.normal(170, 20, 1000),
    ]
)

np.random.shuffle(feat)

df = pd.DataFrame({"feat": feat})
df.head(10)
```

```python
nbins = 3
bin_labels = ["bin_" + str(i) for i in range(1, 4)]
breaks = jenkspy.jenks_breaks(df["feat"], nb_class=nbins)
breaks
```

```python
df["feat_bin"] = pd.cut(df["feat"], bins=breaks, labels=bin_labels)
df.head(10)
```

```python
plt.subplot(1, 2, 1)
hist = plt.hist(df["feat"], bins=20, align="left", color="tab:blue", edgecolor="black")
for b in breaks:
    plt.vlines(b, ymin=0, ymax=max(hist[0]), color="tab:red")
plt.xlabel("Feature")
plt.ylabel("Count")
plt.subplot(1, 2, 2)
df["feat_bin"].value_counts().plot(kind="bar", rot=0, edgecolor="black")
plt.xlabel("Feature Bin")
plt.ylabel("Count")
plt.tight_layout()
```
