```python
# df = pd.DataFrame({"age": np.random.randint(0, 100, 1000)})
df = pd.DataFrame({"age": np.random.normal(50, 10, 1000)})
df.head(10)
```

```python
quantile_list = [0, 0.25, 0.5, 0.75, 1.0]
quantiles = df["age"].quantile(quantile_list)

df["age_quantile"] = pd.qcut(
    df["age"], quantile_list, labels=["1st", "2nd", "3rd", "4th"]
)

print(quantiles)
```

```python
df.head(10)
```

```python
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.hist(df["age"], bins=20, edgecolor="black")
plt.xlabel("Listen Count")
plt.ylabel("Count")
plt.subplot(2, 2, 2)
plt.boxplot(df["age"].values)
plt.xlabel("Listen Count")
plt.ylabel("Count")
plt.subplot(2, 2, 3)
plt.hist(df["age"], bins=20, edgecolor="black")
for quantile in quantiles:
    plt.axvline(quantile, color="r")
plt.subplot(2, 2, 4)
df["age_quantile"].value_counts().plot(kind="bar", rot=0, edgecolor="black")
plt.xlabel("Listen")
plt.ylabel("Count")
plt.tight_layout()
```
