````python
import random
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
````

````python
plt.rcParams["figure.figsize"] = (12, 4)
# plt.rcParams["axes.grid"] = True
plt.rcParams.update({"font.size": 14})
````

````python
df = pd.DataFrame({"months": np.random.randint(0, 120, 1000)})
df.head(10)
````

#### Fixed width binning

``` 
Months Range : Bin
---------------
00 to 12  : 1
13 to 24  : 2
25 to 36  : 3
37 to 48  : 4
49 to 60  : 5
61 to 72  : 6
.............
... and so on
```

````python
bin_width = 12.0
````

````python
df["months_bin_floor"] = np.floor(df["months"] / bin_width).astype("Int64")
df["months_bin_ceil"] = np.ceil(df["months"] / bin_width).astype("Int64")
df.head(10)
````

````python
plt.subplot(1, 2, 1)
plt.hist(df["months"], bins=20, edgecolor="black")
plt.xlabel("Months")
plt.ylabel("Count")
plt.subplot(1, 2, 2)
df["months_bin_floor"].value_counts().plot(kind="bar", rot=0, edgecolor="black")
plt.xlabel("Months Bin")
plt.ylabel("Count")
````

#### Fixed number of bins


````python
nbins = 5
````

````python
# width = (max - min )/ nbins
width = round((df["months"].max() - df["months"].min()) / nbins)
width
````

````python
df["months_bin_floor"] = np.floor(df["months"] / width).astype("Int64")
df["months_bin_ceil"] = np.ceil(df["months"] / width).astype("Int64")

df["months_bin_cut"] = pd.cut(df["months"], bins=nbins, labels=False)
df["months_bin_label"] = pd.cut(df["months"], bins=nbins)
df["months_bin_label_2"] = pd.cut(df["months"], bins=[0, 24, 48, 72, 96, 120])

df.head(10)
````

````python
plt.subplot(1, 2, 1)
plt.hist(df["months"], bins=20, edgecolor="black")
plt.xlabel("months")
plt.ylabel("Count")
plt.subplot(1, 2, 2)
df["months_bin_floor"].value_counts().plot(kind="bar", rot=0, edgecolor="black")
plt.xlabel("Months Bin")
plt.ylabel("Count")

# _ = plt.hist(df["Age_bin_ceil"], edgecolor="black")
# plt.grid()
````
