```python
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (12, 4)
# plt.rcParams["axes.grid"] = True
plt.rcParams.update({"font.size": 14})
```

```python
df = pd.DataFrame({"rating": np.random.random(1000)})
df.head(10)
```

```python
df['rating_10'] = np.round(df['rating']*10).astype('int')
df['rating_100'] = np.round(df['rating']*100).astype('int')
df.head(10)
```

```python
plt.subplot(1, 3, 1)
plt.hist(df["rating"], bins=10, edgecolor="black")
plt.xlabel("Listen Count")
plt.ylabel("Count")
plt.subplot(1, 3, 2)
plt.hist(df["rating_100"], bins=10, edgecolor="black")
plt.xlabel("Listen Count")
plt.ylabel("Count")
plt.subplot(1, 3, 3)
plt.hist(df["rating_10"], bins=10, edgecolor="black")
plt.xlabel("Listen")
plt.ylabel("Count")
plt.tight_layout()
```
